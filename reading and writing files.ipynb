```python
"""
For reading an excel file the read_excel function from pandas is used. 

Excel is a very good format of storing data as multiple sheets can belong to a single workbook. The default value parsed into 
the attribute sheet_name is the first sheet. If the workbook as several worksheet and the second, third or nth 
workbook want to be called the name of the workbook need to be fed or parsed into the attribute "sheet_name".
"""
```
